##Download the Model
Use any model on Colab (or any system with >30GB VRAM on your own machine) to load this in f16. If unavailable, use a GPU with minimum 8GB VRAM to load this in 8bit, or with minimum 5GB of VRAM to load in 4bit.

This step can take around 5 minutes the first time. So please be patient :)

```python
model_name = "defog/sqlcoder-7b-2"
tokenizer = AutoTokenizer.from_pretrained(model_name)
if available_memory > 15e9:
    # if you have atleast 15GB of GPU memory, run load the model in float16
    model = AutoModelForCausalLM.from_pretrained(
        model_name,
        trust_remote_code=True,
        torch_dtype=torch.float16,
        device_map="auto",
        use_cache=True,
    )
else:
    # else, load in 8 bits – this is a bit slower
    model = AutoModelForCausalLM.from_pretrained(
        model_name,
        trust_remote_code=True,
        # torch_dtype=torch.float16,
        load_in_8bit=True,
        device_map="auto",
        use_cache=True,
    )
```

##Set the Question & Prompt and Tokenize
Feel free to change the schema in the prompt below to your own schema

```python
prompt = """### Task
Generate a SQL query to answer [QUESTION]{question}[/QUESTION]

### Instructions
- If you cannot answer the question with the available database schema, return 'I do not know'
- Remember that revenue is price multiplied by quantity
- Remember that cost is supply_price multiplied by quantity

### Database Schema
This query will run on a database whose schema is represented in this string:
CREATE TABLE products (
  product_id INTEGER PRIMARY KEY, -- Unique ID for each product
  name VARCHAR(50), -- Name of the product
  price DECIMAL(10,2), -- Price of each unit of the product
  quantity INTEGER  -- Current quantity in stock
);

CREATE TABLE customers (
   customer_id INTEGER PRIMARY KEY, -- Unique ID for each customer
   name VARCHAR(50), -- Name of the customer
   address VARCHAR(100) -- Mailing address of the customer
);

CREATE TABLE salespeople (
  salesperson_id INTEGER PRIMARY KEY, -- Unique ID for each salesperson
  name VARCHAR(50), -- Name of the salesperson
  region VARCHAR(50) -- Geographic sales region
);

CREATE TABLE sales (
  sale_id INTEGER PRIMARY KEY, -- Unique ID for each sale
  product_id INTEGER, -- ID of product sold
  customer_id INTEGER,  -- ID of customer who made purchase
  salesperson_id INTEGER, -- ID of salesperson who made the sale
  sale_date DATE, -- Date the sale occurred
  quantity INTEGER -- Quantity of product sold
);

CREATE TABLE product_suppliers (
  supplier_id INTEGER PRIMARY KEY, -- Unique ID for each supplier
  product_id INTEGER, -- Product ID supplied
  supply_price DECIMAL(10,2) -- Unit price charged by supplier
);

-- sales.product_id can be joined with products.product_id
-- sales.customer_id can be joined with customers.customer_id
-- sales.salesperson_id can be joined with salespeople.salesperson_id
-- product_suppliers.product_id can be joined with products.product_id

### Answer
Given the database schema, here is the SQL query that answers [QUESTION]{question}[/QUESTION]
[SQL]
"""
```

##Generate the SQL
This can be excruciatingly slow on a T4 in Colab, and can take 10-20 seconds per query. On faster GPUs, this will take ~1-2 seconds

Ideally, you should use `num_beams`=4 for best results. But because of memory constraints, we will stick to just 1 for now.

```python
import sqlparse

def generate_query(question):
    updated_prompt = prompt.format(question=question)
    inputs = tokenizer(updated_prompt, return_tensors="pt").to("cuda")
    generated_ids = model.generate(
        **inputs,
        num_return_sequences=1,
        eos_token_id=tokenizer.eos_token_id,
        pad_token_id=tokenizer.eos_token_id,
        max_new_tokens=400,
        do_sample=False,
        num_beams=1,
    )
    outputs = tokenizer.batch_decode(generated_ids, skip_special_tokens=True)

    torch.cuda.empty_cache()
    torch.cuda.synchronize()
    # empty cache so that you do generate more results w/o memory crashing
    # particularly important on Colab – memory management is much more straightforward
    # when running on an inference service
    return sqlparse.format(outputs[0].split("[SQL]")[-1], reindent=True)
```

```python
question = "What was our revenue by product in the New York region last month?"
generated_sql = generate_query(question)
```

```python
print(generated_sql)
```

```python
# 3 creative prompt versions for Text-to-SQL (SQLCoder style)
# Usage:
#   prompt = PROMPT_V1.format(schema=SCHEMA, question=QUESTION)
#   (then send `prompt` to your model)

SCHEMA = """CREATE TABLE products (
  product_id INTEGER PRIMARY KEY,
  name VARCHAR(50),
  price DECIMAL(10,2),
  quantity INTEGER
);

CREATE TABLE customers (
   customer_id INTEGER PRIMARY KEY,
   name VARCHAR(50),
   address VARCHAR(100)
);

CREATE TABLE salespeople (
  salesperson_id INTEGER PRIMARY KEY,
  name VARCHAR(50),
  region VARCHAR(50)
);

CREATE TABLE sales (
  sale_id INTEGER PRIMARY KEY,
  product_id INTEGER,
  customer_id INTEGER,
  salesperson_id INTEGER,
  sale_date DATE,
  quantity INTEGER
);

CREATE TABLE product_suppliers (
  supplier_id INTEGER PRIMARY KEY,
  product_id INTEGER,
  supply_price DECIMAL(10,2)
);

-- joins:
-- sales.product_id -> products.product_id
-- sales.customer_id -> customers.customer_id
-- sales.salesperson_id -> salespeople.salesperson_id
-- product_suppliers.product_id -> products.product_id
"""

QUESTION = "What was our revenue by product in the New York region last month?"


# -------------------------
# Version 1: "Exam mode" (strict rubric + single SQL)
# -------------------------
PROMPT_V1 = """### EXAM: Text-to-SQL
You must output ONE valid SQL query or exactly: I do not know

### GRADING RUBRIC (follow strictly)
- Use ONLY tables/columns in the schema.
- Prefer explicit JOIN ... ON ...
- No commentary, no markdown, no backticks.
- revenue = SUM(products.price * sales.quantity)
- cost    = SUM(product_suppliers.supply_price * sales.quantity)
- If the question requires data not in schema, output: I do not know

### SCHEMA
{schema}

### QUESTION
{question}

SQL:
"""


# -------------------------
# Version 2: "SQL Linter Cop" (forbids hallucinations + forces auditability)
# -------------------------
PROMPT_V2 = """You are SQL Linter Cop. Your job: generate correct SQL and avoid hallucinations.

HARD RULES:
1) You may only reference these tables: products, customers, salespeople, sales, product_suppliers.
2) You may only reference columns that appear in the schema.
3) If you need anything else, output exactly: I do not know
4) Always include table aliases and qualify columns (e.g., s.sale_date).
5) If a time period is mentioned:
   - "last month" = previous calendar month relative to CURRENT_DATE
   - "last N days" = sale_date >= CURRENT_DATE - INTERVAL 'N days'
6) Output ONLY SQL.

SCHEMA:
{schema}

QUESTION:
{question}

SQL:
"""


# -------------------------
# Version 3: "Planner → SQL" (forces reasoning steps but outputs SQL only)
# -------------------------
PROMPT_V3 = """### Role
You are a senior analytics engineer. Internally, plan your joins/filters/aggregations.

### Internal Plan (DO NOT PRINT THE PLAN)
- Identify required tables and join keys using schema comments.
- Identify measures:
  * revenue = SUM(p.price * s.quantity)
  * cost    = SUM(ps.supply_price * s.quantity)
- Identify filters (region, dates, product/customer/salesperson).
- Identify grouping (e.g., "by product" => GROUP BY p.name).

### Output Contract
Return ONLY the final SQL query.
If impossible with schema, return exactly: I do not know

### Schema
{schema}

### Question
{question}

SQL:
"""


# Example: build the three full prompts
p1 = PROMPT_V1.format(schema=SCHEMA, question=QUESTION)
p2 = PROMPT_V2.format(schema=SCHEMA, question=QUESTION)
p3 = PROMPT_V3.format(schema=SCHEMA, question=QUESTION)

print("=== V1 ===\n", p1)
print("=== V2 ===\n", p2)
print("=== V3 ===\n", p3)
```

Prompt engineering significantly affects SQL generation reliability. The most robust approach combines:

Explicit schema constraints

Defined metric formulas

Clear output contracts

Explicit handling of ambiguous constructs (especially dates)
